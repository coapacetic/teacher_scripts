```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import math
import re
import glob
import math
import xlsxwriter as eW
```

```python
# Import Spreadsheets into DataFrames

# Import Student Scores
student_Scores = pd.read_csv(r'C:\Users\kyled\Dropbox\1819 WPHS Assessment\SAT Score Scaling\1920_SATPracticeTest_Winter_ScaledScores.csv')
student_Scores.head()

# Create Math Scores DataFrame
math_Scores = student_Scores[student_Scores['Subject'] == 'Math']

# Create English Scores DataFrame
english_Scores = student_Scores[student_Scores['Subject'] == 'ELA']

# Import Math Conversion
math_Scale = pd.read_excel(r'C:\Users\kyled\Dropbox\1819 WPHS Assessment\SAT Score Scaling\1920 SAT Diagnostic Curve_MATH.xlsx')
math_Scale.index =  math_Scale.loc[:,'SAT Score']
math_Scale.head()

# Import English Conversion
english_Scale = pd.read_excel(r'C:\Users\kyled\Dropbox\1819 WPHS Assessment\SAT Score Scaling\1920 SAT Diagnostic Curve_ENGLISH.xlsx')
english_Scale.index = english_Scale.loc[:,'SAT Score']

student_Scores.head()
```

```python
# For each student look up, Math Gradebook Score
for n in math_Scores.index:
    if math_Scores.at[n, 'Grade'] == 11:
        mathSAT = math_Scores.at[n, 'Scaled Score']
        scaledGrade = math_Scale.at[mathSAT, 'Grade 11']
        math_Scores.at[n, 'mathGr'] = scaledGrade
      
    elif math_Scores.at[n, 'Grade'] == 10:
        mathSAT = math_Scores.at[n, 'Scaled Score']
        scaledGrade = math_Scale.at[mathSAT, 'Grade 10']
        math_Scores.at[n, 'mathGr'] = scaledGrade
        
    elif math_Scores.at[n, 'Grade'] == 9:
        mathSAT = math_Scores.at[n, 'Scaled Score']
        scaledGrade = math_Scale.at[mathSAT, 'Grade 9']
        math_Scores.at[n, 'mathGr'] = scaledGrade
        
    else:
        math_Scores.at[n, 'mathGr'] = 'Error'

math_Scores.head()

```

```python
# Look up English Gradebook Score

for n in english_Scores.index:
    if english_Scores.at[n, 'Grade'] == 11:
        englishSAT = english_Scores.at[n, 'Scaled Score']
        scaledGrade = english_Scale.at[englishSAT, 'Grade 11']
        english_Scores.at[n, 'engGr'] = scaledGrade
      
    elif english_Scores.at[n, 'Grade'] == 10:
        englishSAT = english_Scores.at[n, 'Scaled Score']
        scaledGrade = english_Scale.at[englishSAT, 'Grade 10']
        english_Scores.at[n, 'engGr'] = scaledGrade
        
    elif english_Scores.at[n, 'Grade'] == 9:
        englishSAT = english_Scores.at[n, 'Scaled Score']
        scaledGrade = english_Scale.at[englishSAT, 'Grade 9']
        english_Scores.at[n, 'engGr'] = scaledGrade
        
    else:
        english_Scores.at[n, 'engGr'] = 'Error'

english_Scores.head()
```

```python
# Export Final Math Frame
math_final = math_Scores.loc[:,['Student Number', 'Grade', 'Name', 'Scaled Score', 'mathGr']]
math_final = math_final.sort_values(['Grade', 'Name'], ascending = [1,1])
math_final

# Write Math Data Frame to Excel
math_final.to_excel('1920_math_winterSAT_gradebookscores.xlsx')

# Export Final English Frame
english_final = english_Scores.loc[:,['Student Number', 'Grade', 'Name', 'Scaled Score', 'engGr']]
english_final = english_final.sort_values(['Grade', 'Name'], ascending = [1,1])
english_final

# Write Math Data Frame to Excel
english_final.to_excel('1920_english_winterSAT_gradebookscores.xlsx')


```
